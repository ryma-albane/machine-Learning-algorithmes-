```python
import pandas as pd                                       
data = pd.read_csv("Salaire_données.csv")          
data.head()
```

```python
x = data[['Experience']]               
y = data['Salaire']
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, color='blueviolet' ,label='Données')
plt.title('visualiser le rapport entre le salaire et l experience ')
plt.xlabel('Annees Experience')
plt.ylabel('Salaire')
plt.legend()
plt.show()  
```

```python
from sklearn.model_selection import train_test_split                      
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train) 
```

```python
import matplotlib.pyplot as plt
y_pred = model.predict(x_test)
plt.scatter(x,y, color='violet' ,label='Données')
plt.plot(x_test, y_pred, color='cyan', label='droite de la régression linéaire')
plt.title('visualiser le rapport entre le salaire et l experience ')
plt.xlabel('Annees Experience')
plt.ylabel('Salaire')
plt.legend()
plt.show()  
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"Score R² de la régression linéaire : {r2}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2)
x_train_poly = poly_features.fit_transform(x_train)
x_test_poly = poly_features.transform(x_test)
```

```python
model = LinearRegression()
model.fit(x_train_poly, y_train)
```

```python
y_pred_poly = model.predict(x_test_poly)    
```

```python
r2 = r2_score(y_test, y_pred_poly)
print(f"Score R² de la régression linéaire : {r2}")
```

```python
X_poly = poly_features.transform(x)
y_pred_poly = model.predict(X_poly)
plt.scatter(x,y, color='salmon', label='Données')
plt.plot(x,y_pred_poly ,color='black', label='courbe de régression polynomiale')
plt.title('visualiser le rapport entre le salaire et l experience ')
plt.xlabel('Annees Experience')
plt.ylabel('Salaire')
plt.legend()
plt.show()  
```

```python
exp = poly_features.transform([[4]])
sal = model.predict(exp)
print(f"salaire pour 4 ans:  ", sal[0])
```
